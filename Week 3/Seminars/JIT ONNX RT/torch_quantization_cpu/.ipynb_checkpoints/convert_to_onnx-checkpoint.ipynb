### Продолжаем предидущий семинар


Теперь мы уже знакомы с ONNX и попробуем дальше ускорить нашу модель с ONNX

```python
import torch
import torch.onnx
import onnx
import torch.onnx
import numpy as np
from time import time
```

```python
def test_time(model, n=100):

    time_total = np.zeros(shape=n)
    # run inference
    batch = torch.rand(1,3,32,32)
    for i in range(n):
        start = time()
        with torch.no_grad():
            output = model(batch)
        end = time()
        time_total[i] = end - start
        
    print(f'execution time for FP32 model is: {time_total.mean():.4f} +/- {time_total.std():.4f}')
```

```python
loaded_model = torch.jit.load('scriptmodule.pt')

test_time(loaded_model)


torch_input = torch.randn(1, 3, 32, 32)

torch.onnx.export(loaded_model,
                torch_input, 
                'model_q.onnx', 
                do_constant_folding=True,
                input_names = ['input'],   # the model's input names
                output_names = ['output'])
```

```python
import onnxruntime
torch_input = torch.randn(1, 3, 32, 32)

options = onnxruntime.SessionOptions()
options.intra_op_num_threads = 10

#onnx_input = onnx_model.adapt_torch_inputs_to_onnx(torch_input)

print(f"Input length: {len(torch_input)}")


ort_session = onnxruntime.InferenceSession("./model_q.onnx",sess_options=options, providers=['CPUExecutionProvider'])

def to_numpy(tensor):
    return tensor.detach().cpu().numpy() if tensor.requires_grad else tensor.cpu().numpy()

ort_inputs = {ort_session.get_inputs()[0].name: to_numpy(torch_input)}



start_time = time()
for i in range(1000):        
    #execution_time = end_time - start_time
    onnxruntime_outputs = ort_session.run(None, ort_inputs)
end_time = time()
execution_time = end_time - start_time
print(f'elapsed in {execution_time/i}')
```
